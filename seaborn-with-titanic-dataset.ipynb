# I visualized titanic dataset using seaborn

```python
import seaborn as sns
%matplotlib inline
```

```python
# Let's import titanic dataset
Titanic = sns.load_dataset('titanic')
Titanic.head()
```

```python
sns.distplot(Titanic['fare'],bins=20,kde=False,color='blue') # plot a univariate distribution of observations
```

```python
sns.countplot(x="class", data=Titanic) # To see distribution of the classes
```

```python
# now let's try to see distribution of man, woman and child in each class
sns.countplot(x="class", hue='who', data=Titanic)
```

```python
sns.countplot( x='who', data=Titanic) # we can see the total number of man, woman and child.
```

```python
# to be able to visualize which group survived mostly
sns.catplot(x="class", hue="who", col="survived",data=Titanic, kind="count", height=5)
```

```python
 sns.catplot( data=Titanic, x="fare", y="age") # to see relations between fare and age
```

```python
sns.scatterplot(data=Titanic, x="fare", y="age")
```
